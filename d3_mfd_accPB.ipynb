```python
# make dataframe with columns for arrival time (t), intensities (rois), filename and sample
data=pd.DataFrame()
for name, items in file_lists_tc.items():
    for csv in items:
        df=pd.read_csv(csv,encoding="utf16",skiprows=2,header=None)
        colnames = ["t", "ROI1", "ROI2", "ROI3", "ROI4", 
                    "ROI5", "ROI6", "ROI7", "ROI8", "ROI9",
                    "ROI10","ROI11","ROI12","ROI13","ROI14",
                    "ROI15","ROI16","ROI17","ROI18","ROI19","ROI20",
                    "ROI21","ROI22","ROI23","ROI24","ROI25","ROI26",
                    "ROI27","ROI28","ROI29","ROI30", "ROI31","ROI32",
                    "ROI33","ROI34","ROI35","ROI36",
                    "ROI37","ROI38","ROI39","ROI40"]
        df.columns=colnames[0:df.shape[1]]
        df["filename"]=os.path.basename(csv)
        df["sample"]=name
        
        data=data.append(df)
```

```python
data
```

```python
# melt dataframe to have one row per intensity value, with columns to identify what it's for (sample and arrival time)
datamelt=data.melt(id_vars=["t","filename","sample"],var_name="roi",value_name="value")

# remove rows with empty values due to extra rois
datamelt=datamelt.dropna(subset=["value"])
```

```python
# convert lifetime to ns
datamelt["t"]=datamelt["t"].mul(1000000000)
```

```python
datamelt[["condition","repeat","embryo","bleach_no","pre-post"]] = datamelt["sample"].str.split('_',expand = True)
datamelt["roi"]=datamelt["roi"].astype(str)
datamelt["nucleus_id"]=datamelt["condition"]+"_"+datamelt["embryo"]+"_"+datamelt["roi"]
datamelt["bleach_id"]=datamelt["nucleus_id"]+"_"+datamelt["bleach_no"]+"_"+datamelt["pre-post"]
```

```python
datamelt
```

```python
# calculate mean intensity of each nucleus 
datamelt["mean_intensity_per_nucleus"]=datamelt.groupby(["bleach_id"])["value"].transform("mean")

# calculate mean intensity of each embryo
datamelt["mean_intensity_per_embryo"]=datamelt.groupby(["sample"])["value"].transform("mean")
```

```python
# calculate mfd
datamelt["mfd_a"]=datamelt["t"]/2.6
datamelt["mfd_a"]=1-datamelt["mfd_a"]
datamelt["mfd_b"]=datamelt["t"]/5.2
datamelt["mfd_b"]=datamelt["mfd_b"]-1
datamelt["mfd_b"]=datamelt["mfd_b"]**2
datamelt["mfd"]=(datamelt["mfd_a"]/datamelt["mfd_b"])*100
datamelt["mfd"]=np.where(datamelt["mfd"]<0,0,datamelt["mfd"])
datamelt
```

```python
# normalise individual nucleus AATs 
datamelt["normalised_per_nucleus"]=datamelt["value"]/datamelt.groupby(["bleach_id"])["value"].transform("max")
# Average normalised AATs across embryos for each condition + Re-normalise
datamelt["normalised_per_nucleus_mean_over_condition"]=datamelt.groupby(["condition","t","pre-post"])["normalised_per_nucleus"].transform("mean")
datamelt["normalised_per_sample"]=datamelt["normalised_per_nucleus_mean_over_condition"]/datamelt.groupby(["condition","pre-post"])["normalised_per_nucleus_mean_over_condition"].transform("max")
```

```python
# plot average normalised Intensity vs AAT across nulcei for each stage+condition, INDIVIDUAL BIOLOGICAL REPLICATES: 
#make palette
palette = {"pre":"red",
           "post":"blue"}

sns.set(font="Arial",style="ticks")

ax=sns.lineplot(data=datamelt.loc[(datamelt["condition"]=="mcherry-mcitrine")],x="mfd",y="normalised_per_sample", hue="pre-post", palette=palette)

ax.set_xlim([0,100])


ax.set(yticks=[0,0.5,1])
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

ax.set_xlabel("mFd [%]", fontsize=30)
ax.set_ylabel("Intensity",fontsize=30)
```

```python
#calculate lifetime at maximum intensity of each nucleus
datamelt["t_max"]=datamelt.loc[datamelt.groupby(["bleach_id"]).value.idxmax(),"t"]
datamelt["t_max"]=datamelt.groupby(["bleach_id"])["t_max"].transform("max")
```

```python
# calculate mean weighted lifetime

#calculate maximum intensity of each nucleus
datamelt["I_max"]=datamelt.loc[datamelt.groupby(["bleach_id"]).value.idxmax(),"value"]
datamelt["I_max"]=datamelt.groupby(["bleach_id"])["I_max"].transform("max")

#first by weighting intensities per lifetime int + summing across each roi
datamelt["weighted_intensity_per_nucleus"]=datamelt["value"]/datamelt["I_max"]
datamelt["sum_weighted_intensity_per_nucleus"]=datamelt.groupby(["bleach_id"])["weighted_intensity_per_nucleus"].transform("sum")

# then by multiplying weighted intensity * lifetime
datamelt["weighted_intensity*lifetime_per_nucleus"]=datamelt["weighted_intensity_per_nucleus"]*datamelt["t"]
datamelt["sum_weighted_intensity*lifetime_per_nucleus"]=datamelt.groupby(["bleach_id"])["weighted_intensity*lifetime_per_nucleus"].transform("sum")

# mean weighted lifetime per nucleus
datamelt["mean_weighted_lifetime_per_nucleus"]=datamelt["sum_weighted_intensity*lifetime_per_nucleus"]/datamelt["sum_weighted_intensity_per_nucleus"]
```

```python
#average mean weighted lifetime across nuclei for each condition
datamelt["average_mean_weighted_lifetime_per_condition"]=datamelt.groupby(["condition","t"])["mean_weighted_lifetime_per_nucleus"].transform("mean")

#average most prominent lifetime across nuclei for each condition
datamelt["average_t_max_per_condition"]=datamelt.groupby(["condition","t"])["t_max"].transform("mean")
```

```python
#mean weighted mfd

# then by multiplying weighted intensity * lifetime
datamelt["weighted_intensity*mfd_per_nucleus"]=datamelt["weighted_intensity_per_nucleus"]*datamelt["mfd"]
datamelt["sum_weighted_intensity*mfd_per_nucleus"]=datamelt.groupby(["bleach_id"])["weighted_intensity*mfd_per_nucleus"].transform("sum")

# mean weighted lifetime per nucleus
datamelt["mean_weighted_mfd_per_nucleus"]=datamelt["sum_weighted_intensity*mfd_per_nucleus"]/datamelt["sum_weighted_intensity_per_nucleus"]
datamelt["mean_weighted_mfd_per_nucleus"]=np.where(datamelt["mean_weighted_mfd_per_nucleus"]<0,0,datamelt["mean_weighted_mfd_per_nucleus"])

#average wean weighted lifetime across nuclei for each condition
datamelt["average_mean_weighted_mfd_per_condition"]=datamelt.groupby(["condition","mfd"])["mean_weighted_mfd_per_nucleus"].transform("mean")
```

```python
#tidy dataframe - drop nucleus duplicates
dfmelt = datamelt[["nucleus_id","bleach_id","bleach_no","pre-post", "embryo","condition","mean_weighted_lifetime_per_nucleus","I_max","t_max","mean_weighted_mfd_per_nucleus"]].drop_duplicates(subset="bleach_id")
dfmelt["sample"]=dfmelt["condition"]+"_"+dfmelt["embryo"]
dfmelt
```

```python
#calculate pre vs post wmwAAT and t_max per nucleus
dfmelt["pre+post"] = dfmelt.groupby(["nucleus_id","bleach_no"])["mean_weighted_mfd_per_nucleus"].transform("sum")
dfmelt["pre_vs_post"] = dfmelt["pre+post"]-2*dfmelt["mean_weighted_mfd_per_nucleus"]
dfmelt
```

```python
prepost = dfmelt[["bleach_id","condition","embryo","nucleus_id","pre_vs_post","pre-post"]]
prepost = prepost.loc[(prepost["pre-post"]=="pre")]

prepost["sample"]=prepost["condition"]+"_"+prepost["embryo"]
```

```python
# plot mean weighted mfD of individual nuclei pre/ post  bleach
#make palette
palette = { "pre":"red",
           "post":"blue"}

#font for graph
sns.set(font="Arial",style="ticks")

#order of conditions to plot
hue_order = ["pre","post"]
order =["mcherry+mcitrine","mcherry-mcitrine"]

#bar and scatterplots
ax=sns.catplot(data=dfmelt,x="condition",y="mean_weighted_mfd_per_nucleus", hue="pre-post", hue_order=hue_order, order=order, kind="box", showfliers=False, legend=False, height = 6, aspect = 1.3, palette=palette)
ax=sns.stripplot(data=dfmelt,x="condition",y="mean_weighted_mfd_per_nucleus",hue="pre-post", hue_order=hue_order, order=order, marker="o", alpha=0.5, jitter = 0.3, dodge = True , size = 3, linewidth=0.5, palette=palette)

#set x and y axis labels
ax.set_ylabel("Mean Weighted\nmfD [%]",fontsize=40)
ax.set_xlabel("")

# set x axis tick labels
ax.set_xticklabels([ "Co-inj+\nmCh + mCit","Tandem\nmCh-mCit"], fontsize = 30)

#set y axis tick intervals and font size
ax.set_ylim([0,40])
ax.yaxis.set_major_locator(ticker.MultipleLocator(10))
plt.yticks(fontsize=30)


# set legend for pre/post bleach
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[:2], labels=labels[:2], title="", fontsize=20, bbox_to_anchor=(1.02, 1.02), loc='upper right' , borderaxespad=0)
```
