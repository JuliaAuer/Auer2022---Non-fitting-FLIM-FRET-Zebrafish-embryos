```python
# make dataframe with columns for nucleus (roi), type (NA), pixel coordinates (x/y/z), value (aat/int), filename and sample
data=pd.DataFrame()
for name, items in file_lists.items():
    for csv in items:
        df=pd.read_csv(csv,encoding="utf-8",skiprows=1, usecols=range(2,8),header=None)
        colnames = ["roi", "type", "x", 
                    "y", "z", "value"]
        df.columns=colnames
        df["filename"]=os.path.basename(csv)
        df["sample"]=name
        
        data=data.append(df)
```

```python
data
```

```python
#re-jig table to get columns for roi(nucleus), type (NA), embryo, repeat, condition,  bleach no., pre/post bleach, x/y/z (pixel co-ordinates), mcitrine AAT, mcitrine intensity, mcherry intensity
data = data.drop(columns = ["type"])

data[["condition","measure","repeat","embryo","bleach_no","pre-post"]] = data["sample"].str.split('_',expand = True)
dfmelt = pd.pivot(data,columns=["measure"],values=["value"],index=["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z"]).reset_index()
dfmelt.columns = dfmelt.columns.droplevel(0)
dfmelt = dfmelt.rename_axis(None,axis=1)
dfmelt.columns = ["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z","raw_aat","int","mch"]

dfmelt
```

```python
#correct aat with formula
dfmelt["aat"]=(dfmelt["raw_aat"]*0.097)-1

dfmelt=dfmelt.drop(columns=['raw_aat'])

dfmelt
```

```python
# calculate mean intensity of each nucleus 

dfmelt["roi"]=dfmelt["roi"].astype(str)
dfmelt["nucleus_id"]=dfmelt["condition"]+"_"+dfmelt["embryo"]+"_"+dfmelt["roi"]

dfmelt["mean_mch_intensity_per_nucleus"]=dfmelt.groupby(["nucleus_id","bleach_no","pre-post"])["mch"].transform("mean")
dfmelt["mean_cit_intensity_per_nucleus"]=dfmelt.groupby(["nucleus_id","bleach_no","pre-post"])["int"].transform("mean")
```

```python
#bin by 0.1 ns aat
interval_range = pd.interval_range(start=0, freq=0.1, end=5, closed='left')
dfmelt['aat_bins'] = pd.cut(dfmelt['aat'], bins=interval_range)
dfmelt["aat_bins"] = dfmelt["aat_bins"].apply(lambda x: pd.Interval(left=round(x.left,2), right=round(x.right,2)))
dfmelt
```

```python
# sum pixel intensities across each bin
binned = dfmelt[["nucleus_id","int","pre-post","bleach_no","aat_bins"]].groupby(["nucleus_id",
    "bleach_no","pre-post","aat_bins"]).sum("int").reset_index()
binned[["condition","embryo","roi"]]= binned["nucleus_id"].str.split('_',expand = True)
binned["sample"]=binned["condition"]+"_"+binned["embryo"]
binned["bleach_id"]=binned["nucleus_id"]+"_"+binned["bleach_no"]+"_"+binned["pre-post"]

#get bin  midpoints
binned["aat"] = binned["aat_bins"].apply(lambda x: x.mid)
binned = binned.drop(columns = ["aat_bins"])
binned
```

```python
binned = pd.read_csv("d1_bleach_binned.csv", usecols=range(1,14))
```

```python
#calculate lifetime at maximum intensity of each nucleus BINNED
binned["t_max"]=binned.loc[binned.groupby(["bleach_id"]).int.idxmax(),"aat"]
binned["t_max"]=binned.groupby(["bleach_id"])["t_max"].transform("max")
```

```python
# calculate mean weighted lifetime

#calculate maximum intensity of each nucleus
binned["I_max"]=binned.loc[binned.groupby(["bleach_id"]).int.idxmax(),"int"]
binned["I_max"]=binned.groupby(["bleach_id"])["I_max"].transform("max")

#first by weighting intensities per lifetime int + summing across each roi
binned["weighted_intensity_per_nucleus"]=binned["int"]/binned["I_max"]
binned["sum_weighted_intensity_per_nucleus"]=binned.groupby(["bleach_id"])["weighted_intensity_per_nucleus"].transform("sum")

# then by multiplying weighted intensity * lifetime
binned["weighted_intensity*lifetime_per_nucleus"]=binned["weighted_intensity_per_nucleus"]*binned["aat"]
binned["sum_weighted_intensity*lifetime_per_nucleus"]=binned.groupby(["bleach_id"])["weighted_intensity*lifetime_per_nucleus"].transform("sum")

# mean weighted lifetime per nucleus
binned["mean_weighted_lifetime_per_nucleus"]=binned["sum_weighted_intensity*lifetime_per_nucleus"]/binned["sum_weighted_intensity_per_nucleus"]
```

```python
#average mean weighted lifetime across nuclei for each embryo/condition
binned["average_mean_weighted_lifetime_per_embryo"]=binned.groupby(["sample"])["mean_weighted_lifetime_per_nucleus"].transform("mean")
binned["average_mean_weighted_lifetime_per_condition"]=binned.groupby(["condition","aat","pre-post"])["mean_weighted_lifetime_per_nucleus"].transform("mean")

#average most prominent lifetime across nuclei for each embryo/condition
binned["average_t_max_per_embryo"]=binned.groupby(["sample"])["t_max"].transform("mean")
binned["average_t_max_per_condition"]=binned.groupby(["condition","aat","pre-post"])["t_max"].transform("mean")
```

```python
#clean dataframe - drop duplicates
dm = binned[["nucleus_id","bleach_id","roi","condition","embryo","bleach_no","pre-post","mean_weighted_lifetime_per_nucleus","average_mean_weighted_lifetime_per_embryo","average_mean_weighted_lifetime_per_condition","I_max","t_max","average_t_max_per_embryo","average_t_max_per_condition"]].drop_duplicates()
```

```python
#calculate pre vs post wmwAAT and t_max per nucleus
dm["pre+post"] = dm.groupby(["nucleus_id","bleach_no"])["mean_weighted_lifetime_per_nucleus"].transform("sum")
dm["pre_vs_post"] = dm["pre+post"]-2*dm["mean_weighted_lifetime_per_nucleus"]
dm
```

```python
dm_prepost = dm[["bleach_id","condition","embryo","nucleus_id","pre_vs_post","pre-post"]]
dm_prepost = dm_prepost.loc[(dm_prepost["pre-post"]=="pre")]

dm_prepost["sample"]=dm_prepost["condition"]+"_"+dm_prepost["embryo"]
```
