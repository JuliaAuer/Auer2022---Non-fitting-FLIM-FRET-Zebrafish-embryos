```python
# make dataframe with columns for roi(nucleus),type (NA), x/y/z (pixel co-ordinates), value (intensity) filename and sample
data=pd.DataFrame()
for name, items in file_lists.items():
    for csv in items:
        df=pd.read_csv(csv,encoding="utf-8",skiprows=1, usecols=range(2,8),header=None)
        colnames = ["roi", "type", "x", 
                    "y", "z", "value"]
        df.columns=colnames
        df["filename"]=os.path.basename(csv)
        df["sample"]=name
        
        data=data.append(df)
```

```python
data
```

```python
#melt table to get columns for roi, embryo, repeat(replicate), condition, bleach no., pre/post bleach, x/y/z (pixel co-ordinates), mcitrine intensity, mcherry intensity
data = data.drop(columns = ["type"])

data[["condition","measure","repeat","embryo","bleach_no","pre-post"]] = data["sample"].str.split('_',expand = True)
dfmelt = pd.pivot(data,columns=["measure"],values=["value"],index=["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z"]).reset_index()
dfmelt.columns = dfmelt.columns.droplevel(0)
dfmelt = dfmelt.rename_axis(None,axis=1)
dfmelt.columns = ["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z","int","mch"]

dfmelt
```

```python
# calculate mean intensity of each nucleus 

dfmelt["roi"]=dfmelt["roi"].astype(str)
dfmelt["nucleus_id"]=dfmelt["condition"]+"_"+dfmelt["embryo"]+"_"+dfmelt["roi"]
dfmelt["bleach_id"]=dfmelt["nucleus_id"]+"_"+dfmelt["bleach_no"]+"_"+dfmelt["pre-post"]

dfmelt["mean_mch_intensity_per_nucleus"]=dfmelt.groupby(["nucleus_id","bleach_no","pre-post"])["mch"].transform("mean")
dfmelt["mean_cit_intensity_per_nucleus"]=dfmelt.groupby(["nucleus_id","bleach_no","pre-post"])["int"].transform("mean")
```

```python
# clean dataframe - drop duplicates
dfint = dfmelt[["condition","embryo","nucleus_id","bleach_id","bleach_no","pre-post","mean_mch_intensity_per_nucleus","mean_cit_intensity_per_nucleus"]].drop_duplicates("bleach_id")

dfint["sample"]=dfint["condition"]+"_"+dfint["embryo"]
dfint.to_csv("d1_bleach_non-binned_int.csv")
dfint
```

```python
# plot mean mcherry intensity of individual nuclei pre/ post donor bleach
#make palette
palette = { "pre":"red",
           "post":"blue"}

#font for graph
sns.set(font="Arial",style="ticks")

#order of conditions to plot
hue_order = ["pre","post"]
order =["mcherry+mcitrine","mcherry-mcitrine"]

#bar and scatterplots
ax=sns.catplot(data=dfint,x="condition",y="mean_mch_intensity_per_nucleus", hue="pre-post", hue_order=hue_order, order=order, kind="box", showfliers=False, legend=False, height = 6, aspect = 1.3, palette=palette)
ax=sns.stripplot(data=dfint,x="condition",y="mean_mch_intensity_per_nucleus",hue="pre-post", hue_order=hue_order, order=order, marker="o", alpha=0.5, jitter = 0.3, dodge = True , size = 3, linewidth=0.5, palette=palette)

#set x and y axis labels
ax.set_ylabel("Mean Nucleus\nmCherry Intensity",fontsize=30)
ax.set_xlabel("")

# set x axis tick labels
ax.set_xticklabels([ "Co-inj+\nmCh + mCit","Tandem\nmCh-mCit"], fontsize = 25)

#set y axis tick intervals and font size
ax.yaxis.set_major_locator(ticker.MultipleLocator(100))
plt.yticks(fontsize=20)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[:2], labels=labels[:2], title="Bleach")

# set legend for pre/post bleach
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[:2], labels=labels[:2], title="Bleach", fontsize=15, bbox_to_anchor=(1, 1.02), loc='upper left' , borderaxespad=0)

#save plot

plt.savefig("d1_int-FRET_mean_mch_int_prevspost.png",dpi=300, bbox_inches='tight')
```

```python
#calculate apparent FRET efficiency by comparing change in mcit intensity in AccPb per nucleus
dfint["nucleus_bleach"] = dfint["nucleus_id"]+"_"+dfint["bleach_no"]
pre = dfint[["nucleus_id","nucleus_bleach","bleach_no","condition","mean_cit_intensity_per_nucleus"]].loc[(dfint["pre-post"]=="pre")].reset_index()
post = dfint[["nucleus_id","nucleus_bleach","bleach_no","condition","mean_cit_intensity_per_nucleus"]].loc[(dfint["pre-post"]=="post")].reset_index()
merge = pd.merge(pre, post, on="nucleus_bleach")
merge["app_E"]=1-(merge["mean_cit_intensity_per_nucleus_x"]/merge["mean_cit_intensity_per_nucleus_y"])
merge.to_csv("d1_bleach_apparent_fret_eff.csv")
merge
```

```python
#make new clean dataframe with apparent FREt efficiency per nucleus
fret_eff = merge[["nucleus_bleach","app_E"]]
fret_eff[["condition","embryo","roi","bleach_no"]] = fret_eff["nucleus_bleach"].str.split('_',expand = True)
fret_eff["sample"]=fret_eff["condition"]+"_"+fret_eff["embryo"]
fret_eff.to_csv("d1_bleach_apparent_fret_eff_r.csv")
fret_eff
```
