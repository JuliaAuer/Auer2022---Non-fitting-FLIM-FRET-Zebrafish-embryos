```python
# make dataframe with columns for roi(nucleus),type (NA), x/y/z (pixel co-ordinates), value (intensity) filename and sample
data=pd.DataFrame()
for name, items in file_lists.items():
    for csv in items:
        df=pd.read_csv(csv,encoding="utf-8",skiprows=1, usecols=range(2,8),header=None)
        colnames = ["roi", "type", "x", 
                    "y", "z", "value"]
        df.columns=colnames
        df["filename"]=os.path.basename(csv)
        df["sample"]=name
        
        data=data.append(df)
```

```python
data
```

```python
#melt table to get columns for roi, embryo, repeat(replicate), condition, bleach no., pre/post bleach, x/y/z (pixel co-ordinates), mcitrine intensity, mcherry intensity
data = data.drop(columns = ["type"])

data[["condition","measure","repeat","embryo","bleach_no","pre-post"]] = data["sample"].str.split('_',expand = True)
dfmelt = pd.pivot(data,columns=["measure"],values=["value"],index=["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z"]).reset_index()
dfmelt.columns = dfmelt.columns.droplevel(0)
dfmelt = dfmelt.rename_axis(None,axis=1)
dfmelt.columns = ["roi","embryo","repeat","condition","bleach_no","pre-post","x","y","z","cit","int"]

dfmelt
```

```python
# calculate mean intensity of each nucleus 

dfmelt["roi"]=dfmelt["roi"].astype(str)
dfmelt["nucleus_id"]=dfmelt["condition"]+"_"+dfmelt["embryo"]+"_"+dfmelt["roi"]
dfmelt["bleach_id"]=dfmelt["nucleus_id"]+"_"+dfmelt["bleach_no"]+"_"+dfmelt["pre-post"]

dfmelt["mean_cit_intensity_per_nucleus"]=dfmelt.groupby(["bleach_id"])["cit"].transform("mean")
dfmelt["mean_mch_intensity_per_nucleus"]=dfmelt.groupby(["bleach_id"])["int"].transform("mean")
```

```python
# clean dataframe - drop duplicates
dfint = dfmelt[["condition","nucleus_id","embryo","bleach_id","pre-post","bleach_no","mean_cit_intensity_per_nucleus","mean_mch_intensity_per_nucleus"]].drop_duplicates("bleach_id")
dfint.to_csv("e3_bleach_non-binned_ints.csv")
dfint
```

```python
# plot mean mcherry intensity of individual nuclei pre/ post donor bleach
#make palette
palette = { "pre":"red",
           "post":"blue"}

#font for graph
sns.set(font="Arial",style="ticks")

#order of conditions to plot
hue_order = ["pre","post"]

#bar and scatterplots
ax=sns.catplot(data=dfint.drop_duplicates(subset=["bleach_id","mean_mch_intensity_per_nucleus"]),x="condition",y="mean_mch_intensity_per_nucleus", hue="pre-post", hue_order=hue_order, kind="box", showfliers=False, legend=False, height = 6, aspect = 1.3, palette=palette)
ax=sns.stripplot(data=dfint.drop_duplicates(subset=["bleach_id","mean_mch_intensity_per_nucleus"]),x="condition",y="mean_mch_intensity_per_nucleus", hue="pre-post", hue_order=hue_order, marker="o", alpha=0.5, jitter = 0.3, dodge = True , size = 3, linewidth=0.5, palette=palette)

#set x and y axis labels
ax.set_ylabel("Mean Nucleus\nmCherry Intensity",fontsize=30)
ax.set_xlabel("")

#set y axis tick intervals and font size
ax.yaxis.set_major_locator(ticker.MultipleLocator(100))
plt.yticks(fontsize=20)

# set x axis tick labels
ax.set_xticklabels([ "Co-inj+\nmCh + mCit","Tandem\nmCh-mCit"], fontsize = 25)


# set legend for pre/post bleach
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[:2], labels=labels[:2], title="Bleach", fontsize=15, bbox_to_anchor=(1, 1.02), loc='upper left' , borderaxespad=0)

#save plot

plt.savefig("e3_SE_donor-bleach_mean_mch_int_515nm_prevspost.png",dpi=300, bbox_inches='tight')
```

```python
# calculate mch intensity difference pre vs post donor bleaching per nucleus
dfint["pre+post"] = dfint.groupby(["nucleus_id","bleach_no"])["mean_mch_intensity_per_nucleus"].transform("sum")
dfint["pre_vs_post"] = dfint["pre+post"]-2*dfint["mean_mch_intensity_per_nucleus"]
dfint
```

```python
# calculate mcit intensity difference pre vs post donor bleaching per nucleus
dfint["pre+post_cit"] = dfint.groupby(["nucleus_id","bleach_no"])["mean_cit_intensity_per_nucleus"].transform("sum")
dfint["pre_vs_post_cit"] = dfint["pre+post_cit"]-2*dfint["mean_cit_intensity_per_nucleus"]
dfint
```

```python
#make new dataframe to compare pre vs post mcherry intensity pre nucleus
prepost = dfint[["nucleus_id","bleach_id","condition","pre_vs_post","pre-post","embryo","pre_vs_post_cit"]]
prepost = prepost.loc[(prepost["pre-post"]=="pre")]
prepost["sample"]=prepost["condition"]+"_"+prepost["embryo"]
prepost.to_csv("e3_bleach_prevspost_int.csv")
prepost
```
